```python
from nilearn import plotting
from nilearn.plotting import plot_stat_map, plot_anat, plot_img
from nilearn.image import concat_imgs, mean_img
```

```python
PATH='abide.cc.bvol-2.20150118/Caltech/0051456/session_1/anat_1/Caltech_0051456_brain_seg.nii.gz'

plotting.plot_img(PATH)
```

```python
import matplotlib.pyplot as plt
plot_stat_map(PATH, threshold=3.0,
              display_mode='z', cut_coords=3, black_bg=True)

plt.show()
```

```python
display = plotting.plot_anat(PATH)
```

```python
plotting.plot_roi(PATH, cmap=plotting.cm.bwr_r,
                  title='64 regions of brain clusters')
```
